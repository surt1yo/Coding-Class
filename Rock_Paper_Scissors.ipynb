```python

import random
while True:
  r="Rock"
  p="Paper"
  s="Scissors"
  l="You Lose!\n---------------------------------"
  w="You win!\n---------------------------------"
  t="It's a tie\n---------------------------------"
  option=["Rock", "Paper","Scissors"]
  uc=input("Choose a option:\nRock\nPaper\nScissors\n")
  cc=random.choice(option)
  print(f"You chose {uc} and the computer chose {cc}\n---------------------------------")
  if uc==cc:
    print(t)
  elif uc==r and cc==s:
    print(w)
  elif uc==p and cc==s:
    print(l)
  elif uc==r and cc==p:
    print(l)
  elif uc==p and cc==r:
    print(w)
  elif uc==s and cc==p:
    print(w)
  elif uc==s and cc==r:
    print(l)
  else:
    raise ValueError("Invalid Choice")


```
