```python
input1 = int(input("How many nights are you staying?: "))
city = input("Which City are you going to?\n--------------------------\nLos Angeles\nCharlotte\nTampa\nNew York\n--------------------------------").title() # Convert input to title case for consistent matching
ride = int(input("How many days will you use the rental car?: "))
spend = float(input("How much money will you spend?: "))
city_costs = {
    "Los Angeles": 1800,
    "Charlotte": 1500,
    "Tampa": 1400,
    "New York": 1900
}
def hotel_cost(nights):
  return input1 * nights
def plane_ride_cost(city):
  return city_costs.get(city, "Invalid City")
def rental_car_cost(days):
  return days * 25.75
def trip_cost(days, city, nights):
  car_cost = rental_car_cost(days)
  plane_cost = plane_ride_cost(city)
  hotel_c = hotel_cost(nights)
  if isinstance(plane_cost, (int, float)):
    return car_cost + plane_cost + hotel_c + spend
  else:
    return "Invalid city selected for plane ride cost calculation."
print(f"The price of the rental car is ${rental_car_cost(ride):.2f}.")
print(f"The total trip cost is: ${trip_cost(ride, city, input1):.2f}")
```
