```python
import random
play=True
num=str(random.randint(10,20))
print("I will pick a number between 10 and 20. You have to guess the number 1 digit at a time.")
print("The game ends when you get 1 point.")
while play:
  guess=input("Give me your guess: ")
  if guess == num:
    print("You Win!")
    print(f"The number was {num}!")
    break
  else:
    print("You Lose!")
    print(f"The number was {num}!")
    break
```
