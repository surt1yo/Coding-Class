```python
import math
angle = float(input("Enter the angle: "))
unit = input("Degrees? (yes/no): ").strip().lower()
if unit == "yes":
    angle = math.radians(angle)
print("Sine =", f"{math.sin(angle)}")
print("Cosine =", f"{math.cos(angle)}")
try:
    print("Tangent =", f"{math.tan(angle)}")
except:
    print("Tangent is undefined")
```
