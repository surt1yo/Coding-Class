```python
import random
import time
def randomtime(start,end):
  print(f"Printing random date between {start} and {end}:")
  randomgen=random.random()
  datefm="%m/%d/%y"
  starttime=time.mktime(time.strptime(start,datefm))
  endtime=time.mktime(time.strptime(end,datefm))
  randomtime=starttime+randomgen*(endtime-starttime)
  randomdate=time.strftime(datefm,time.localtime(randomtime))
  return(randomdate)
print(randomtime("01/01/20","12/31/25"))
```
