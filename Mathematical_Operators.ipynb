```python
import math

num = float(input("Enter a number for ceil and floor: "))
print(f"The ceiling and floor values of {num} are: {math.ceil(num)}, {math.floor(num)}")

x = float(input("Enter value for x: "))
y = float(input("Enter value for y: "))
print(f"The value of x after copying the sign from y is: {math.copysign(x, y)}")

a = float(input("Enter first number for fabs and gcd: "))
b = float(input("Enter second number for fabs and gcd: "))
print(f"Absolute values of {a} and {b} are: {math.fabs(a)}, {math.fabs(b)}")
print(f"The GCD of {int(a)} and {int(b)} is: {math.gcd(int(a), int(b))}")
```
